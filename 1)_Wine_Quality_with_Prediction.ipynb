```python
import pandas as pd
import numpy as np
```

```python
a = pd.read_csv("/content/drive/MyDrive/ML Datasets/wine-quality.csv")
df = pd.DataFrame(a)
```

```python
df.isnull().sum()
```

```python
df["quality"].value_counts()
```

```python
# 345678
df["quality"] = df["quality"].replace({3,4,5},"Bad")
```

```python
df["quality"] = df["quality"].replace({6,7,8},"Good")
```

```python
df
```

```python
df["quality"].value_counts()
```

```python
df["quality"] = df["quality"].replace("Good",1)
df["quality"] = df["quality"].replace("Bad",0)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
```

```python
X = df.drop(columns="quality").values
```

```python
Y = df["quality"].values
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2, stratify=Y, shuffle=True)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
model = RandomForestClassifier()
```

```python
model.fit(X_train, Y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
X_train_prediction = model.predict(X_train)
X_train_Accuracyscore = accuracy_score(X_train_prediction, Y_train)
print(X_train_Accuracyscore)
```

```python
X_test_prediction = model.predict(X_test)
X_test_Accuracyscore = accuracy_score(X_test_prediction, Y_test)
print(X_test_Accuracyscore)
```

```python
import gradio as gr
import joblib
import numpy as np

# Load the trained model
model = joblib.load("SVM_Model.pkl")

# Define prediction function
def predict_wine_quality(fixed_acidity, volatile_acidity, citric_acid, residual_sugar,
                         chlorides, free_sulfur_dioxide, total_sulfur_dioxide,
                         density, pH, sulphates, alcohol):

    input_data = np.array([[fixed_acidity, volatile_acidity, citric_acid, residual_sugar,
                            chlorides, free_sulfur_dioxide, total_sulfur_dioxide,
                            density, pH, sulphates, alcohol]])

    prediction = model.predict(input_data)[0]
    return "🍷 Good Quality Wine" if prediction == 1 else "⚠️ Bad Quality Wine"

# Define the interface
interface = gr.Interface(
    fn=predict_wine_quality,
    inputs=[
        gr.Number(label="Fixed Acidity"),
        gr.Number(label="Volatile Acidity"),
        gr.Number(label="Citric Acid"),
        gr.Number(label="Residual Sugar"),
        gr.Number(label="Chlorides"),
        gr.Number(label="Free Sulfur Dioxide"),
        gr.Number(label="Total Sulfur Dioxide"),
        gr.Number(label="Density"),
        gr.Number(label="pH"),
        gr.Number(label="Sulphates"),
        gr.Number(label="Alcohol")
    ],
    outputs=gr.Text(label="Prediction"),
    title="🍇 Wine Quality Prediction",
    description="Enter wine chemistry values to predict whether it's good or bad quality.",
)

# Launch the app
interface.launch()
```
